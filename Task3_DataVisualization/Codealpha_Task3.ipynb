```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Hp\Downloads\cleaned_books_data.csv")

df.head()
```

```python
plt.figure(figsize=(6,4))

df["Rating"].value_counts().plot(kind="bar")

plt.title("Book Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.hist(df["Price"])

plt.title("Book Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.barplot(x="Rating", y="Price", data=df)

plt.title("Average Price by Rating")

plt.show()
```

```python
df["Rating"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Rating Percentage Distribution")

plt.ylabel("")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(y=df["Price"])

plt.title("Price Spread Analysis")

plt.show()
```

```python
df.to_csv("cleaned_books_data.csv", index=False)
```
